```python
import time
from huggingface_hub import hf_hub_download
from llama_cpp import Llama

GGUF_REPO_ID = 'your-username/mistral7b-finance-gguf'  # <- your GGUF repo
GGUF_FILENAME = 'model-Q4_K_M.gguf'
path = hf_hub_download(repo_id=GGUF_REPO_ID, filename=GGUF_FILENAME)
llm = Llama(model_path=path, n_ctx=2048, verbose=False)
```

```python
prompt = 'Explain dollar-cost averaging in one paragraph.'
t0 = time.time()
out = llm.create_chat_completion(messages=[{'role': 'user', 'content': prompt}], max_tokens=200)
elapsed = time.time() - t0
text = out['choices'][0]['message']['content']
tok = out.get('usage', {}).get('completion_tokens', 0)
print(text)
# tokens/sec MEASURED live on this CPU — not assumed.
print(f'\n[{tok} tokens in {elapsed:.2f}s = {tok/elapsed:.1f} tok/s on CPU]')
```
